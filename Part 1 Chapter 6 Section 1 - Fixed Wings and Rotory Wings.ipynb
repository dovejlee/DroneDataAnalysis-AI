```python
import pandas as pd
import numpy as np
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
FW_df = pd.read_csv("data/FixedWing.csv")
```

```python
FW_df.head()
```

```python
RW_df = pd.read_csv("data/RotaryWing.csv")
```

```python
RW_df.tail()
```

```python
FW_df["Type"]="FixedWing"
```

```python
FW_df.head()
```

```python
RW_df["Type"]="RotaryWing"
```

```python
UAV_df = pd.concat([FW_df, RW_df])
```

```python
print(UAV_df.iloc[[29,30,0,1]])
```

```python
UAV_df =UAV_df.reset_index(drop=True)
```

```python
UAV_df["diagonal"]= np.sqrt(UAV_df["Length"]**2 + UAV_df["Width"]**2)
```

```python
FixedWing_df = UAV_df[UAV_df["Type"]=="FixedWing"]
RotoryWing_df = UAV_df[UAV_df["Type"]=="RotaryWing"]
```

```python
FixedWing_df
```

```python
RotoryWing_df
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,1,1)
axes2 = fig.add_subplot(2,1,2)
axes1.plot(FixedWing_df["Length"], FixedWing_df["Weight"], 'o',c ="red")
axes2.plot(RotoryWing_df["Length"], RotoryWing_df["Weight"], 'o', c ='blue' )
```

```python
axes1.set_title("Fixed Wing")
axes2.set_title("Rotory Wing")
axes2.set_xlim(axes1.get_xlim())
fig.suptitle("UAV Size <Length-Weight>")
fig
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,1,1)
axes2 = fig.add_subplot(2,1,2)
axes1.plot(FixedWing_df["Length"], FixedWing_df["Width"], 'o',c ="red")
axes2.plot(RotoryWing_df["Length"], RotoryWing_df["Width"], 'o', c ='blue' )
axes1.set_title("Fixed Wing")
axes2.set_title("Rotory Wing")
fig.suptitle("UAV Size <Length-Width>")
axes2.set_xlim(axes1.get_xlim())
axes2.set_ylim(axes1.get_ylim())
fig.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,1,1)
axes2 = fig.add_subplot(2,1,2)
axes1.plot(FixedWing_df["Width"], FixedWing_df["Length"], 'o',c ="red")
axes2.plot(RotoryWing_df["Width"], RotoryWing_df["Length"], 'o', c ='blue' )
axes1.set_title("Fixed Wing")
axes2.set_title("Rotory Wing")
fig.suptitle("UAV Size <Width - Length>")
axes2.set_xlim(axes1.get_xlim())
axes2.set_ylim(axes1.get_ylim())
fig.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,1,1)
axes2 = fig.add_subplot(2,1,2)
axes1.plot(FixedWing_df["diagonal"], FixedWing_df["Weight"], 'o',c ="red")
axes2.plot(RotoryWing_df["diagonal"], RotoryWing_df["Weight"], 'o', c ='blue' )
axes1.set_title("Fixed Wing")
axes2.set_title("Rotory Wing")
fig.suptitle("UAV Size <Diagonal- Weight>")
axes2.set_xlim(axes1.get_xlim())
axes2.set_ylim(axes1.get_ylim())
fig.tight_layout()
```
